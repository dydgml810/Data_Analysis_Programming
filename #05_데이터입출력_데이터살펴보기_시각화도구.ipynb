- 구글 드라이브에 있는 파일 불러오기 위해 Colab과 연동한 후, DataFrame으로 저장하세요. <br/>
- 파일은 아래 링크에서 다운받아서 본인의 구글 Drive에 저장하세요.<br/>

제목 없음.png


```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive') # 구글 드라이브 연동
base_dir = '/content/drive/MyDrive/' # MyDrive는 구글 드라이브의 '내 드라이브'와 같음, 저장해둔 폴더에 따라 설정달리하기
file_path = base_dir + "경기도인구데이터.csv"

# csv 파일을 불러와 DataFrame에 저장하기,
df = pd.read_csv(file_path, encoding='CP949') # csv 파일의 경우 한글이 포함된 파일을 불러오려면 encoding='CP949' 매개변수를 붙여줘야 인식한다.

# '구분'에 해당되는 열은 데이터로 포함이 안되도록 해당 열을 index로 만들어 줌.
df = df.set_index('구분')

#불러온 데이터의 구조를 확인해보자.
df.head(7)
```

2. 데이터 프레임에서 2010년도에 해당되는 Column만 가져와 해당 Series를 출력해 보세요.

```python
column_2010 = df.loc[:, '2010']
print(column_2010.head(10))
```

## 데이터 살펴보기
앞선 실습에서 파일을 불러와 Dataframe으로 불러왔다.<br/>
수업에서 배운 통계함수를 이용해 최댓값과 평균값을 출력해보자.

```python
max_value = df.max()
mean_value = df.mean()

print('연도별 최대 인구수')
print(max_value)

print('\n연도별 평균 인구수')
print(mean_value)
```

## 데이터 시각화
수업시간에 배운 Matplotlib 도구를 이용하여 양주시의 연도별 인구 변화 그래프를 그려보시오.<br/>
이때, marker 매개변수를 이용하여 데이터에 마커를 표시해보시오.

```python
import matplotlib.pyplot as plt

# 양주시에 해당하는 row를 가져오기 df.loc을 이용
yangju = df.loc['양주시', :]
print(yangju)

# plt 모듈로 그래프 작성하기
plt.figure(figsize=(10,6))

plt.plot(yangju.index, yangju.values, color='b',marker='o') # hint) yangju의 index가 x값, values가 y값
# plt.plot(yangju, color='b', marker='o')

plt.xlabel('Year')
plt.ylabel('Population')

# 그래프를 출력하는 코드
plt.show()
```
